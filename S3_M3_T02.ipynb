### 2. Encuentra cuántos datos faltantes hay por columna.

Para saber cuantos datos faltan por columna, usamos `isna()` que nos dice cuantos valores ausentes hay en cada columna, pero en forma de tabla y dificil de leer al usar un dataset tan grande.  Para solucionarlo, agregamos el método `.sum()`.

```python
df_flights_new.isna().sum()
```

Una vez que sabemos que datos faltan en cada columna, comenzaremos a limpiarlos uno por uno comenzando con los más pequeños.  

Primero crearemos una copia del data frame `df_flights_new` para poder hacer los cambios tranquilamente.

```python
df_flights_copia = df_flights_new.copy()
df_flights_copia
```

En la columna _TailNum_ rellenaremos los datos _NaN_ con **"No disponible"** usando `.fillna()` y después comprobaremos que todo haya salido bien.

```python
df_flights_copia['TailNum'].fillna('No disponible', inplace=True)
df_flights_copia[df_flights_copia['TailNum'] == 'No disponible']
```

```python
df_flights_copia['TailNum'].isna().sum()
```

La siguiente columna con menos valores vacíos (198), es _CRSElapsedTime_.  Nos dimos cuenta que esta columna también se relaciona con _ActualElapsedTime_ y con _AirTime_, pero por si solas no nos dicen nada pues coinciden en los espacios vacios. Es decir, las valores vacíos de _CRSElapsedTime_ son los mismos que en _ActualElapsedTime_, _ArrDelay_ y con _AirTime_.

Al  hacer un análisis de todas las columnas, se llegó a la conclusión que lo que tenían en común era la columna de _Distance_ y a partir de ahí se pudo sacar la media de estas columnas usando `.groupby()` y `mean()`.

```python
means = df_flights_copia.groupby('Distance')[['CRSElapsedTime', 'ActualElapsedTime', 'ArrDelay', 'AirTime']].mean()
means
```

Teniendo esta información, se pueden rellenar las columnas: 
* _AirTime_
* _ActualElapsedTime_
* _ArrDelay_
* _CRSElapsedTime_

Para lograrlo se usaron `.fillna()` y `map`.

```python
df_flights_copia['AirTime'].fillna(df_flights_copia['Distance'].map(means['AirTime']), inplace=True)
df_flights_copia['ActualElapsedTime'].fillna(df_flights_copia['Distance'].map(means['ActualElapsedTime']), inplace=True)
df_flights_copia['ArrDelay'].fillna(df_flights_copia['Distance'].map(means['ArrDelay']), inplace=True)
df_flights_copia['CRSElapsedTime'].fillna(df_flights_copia['Distance'].map(means['CRSElapsedTime']), inplace=True)
df_flights_copia.isna().sum()
```

Para completar las columnas _ArrTime_, se toma el percentil 75 pues así nos aseguramos de usar los valores más usados.  

Usaremos `.quantile(0.75)` para poder hacerlo.

```python
df_flights_copia['ArrTime'].fillna(df_flights_copia['ArrTime'].quantile(0.75), inplace=True)
df_flights_copia.isna().sum()
```

Finalmente, los datos que no se pudieron completar con media, mediana, percentil, etc., se completaran con `.interpolate` pues son muy pequeños y no afectan en nada a la base de datos.

```python
df_flights_copia['ActualElapsedTime'] = df_flights_copia['ActualElapsedTime'].interpolate()
df_flights_copia['CRSElapsedTime'] = df_flights_copia['CRSElapsedTime'].interpolate()
df_flights_copia['AirTime'] = df_flights_copia['AirTime'].interpolate()
df_flights_copia['ArrDelay'] = df_flights_copia['ArrDelay'].interpolate()
df_flights_copia.isna().sum()
```

### 3. Crea nuevas columnas de Velocidad Media del vuelo y si ha llegado tarde o no.

#### Columna de "Velocidad Media".

Para hacerlo se crea una la columna _'Velocidad_Media'_ dividiendo los valores de _'AirTime_ dividido entre 60 - para convertir los minutos a horas - y entre la Distancia para sacar la velocidad promedio.

```python
df_flights_copia['Velocidad_Media'] = df_flights_copia['Distance'] / (df_flights_copia['AirTime'] / 60)
df_flights_copia
```

#### Columna de "Arribo"
Para hacerlo se usa la columna _Cancelled_ para saber si un vuelo arribó a su destino o no.  Esto se hace usando el método de `.apply()` y la función `lambda`.

```python
df_flights_copia['Arribo'] = df_flights_copia['Cancelled'].apply(lambda x: 'No' if x == 1 else 'Sí')
df_flights_copia
```

### 4. Haz una tabla de las aerolíneas con más retrasos acumulados.

Para hacerlo usamos `.groupby()`,`.sum()` y `.sort_values()` para crear la agrupación de las aerolineas sumando los valores de _'ArrDelay'_ para así saber el número de minutos acumulados en los retrasos. Finalmente se ordenan de mayor a menor y se usa `pd.DataFrame()` para crear el dataframe.

```python
pd.DataFrame(df_flights_copia.groupby('UniqueCarrier')[['ArrDelay','DepDelay']].sum().sort_values(by='ArrDelay', ascending=False).reset_index())
```

### 5.1 ¿Cuáles son los vuelos más largos?

Para hacerlo usamos el método `.sort_values()`con la especificación `ascending=False` y también `.drop_duplicates()` para evitar que  el dataframe fuera tan largo.

```python
vuelos_largos = df_flights_copia[['FlightNum', 'Distance', 'Origin','Dest']].drop_duplicates().sort_values(by='Distance', ascending=False).reset_index(drop=True)
vuelos_largos
```

### 5.2 ¿Y los que tienen más retrasados?

Para esta filtración, también usamos `groupby()` y `sort_values()` como el código anterior, pero ademas el método `sum()` y `.agg()` para agregar las columnas de _Origin_ y _Dest_ y así tener una mejor presentación.

```python
vuelos_delay = df_flights_copia.groupby('FlightNum').agg({'ArrDelay':'sum', 'DepDelay':'sum', 'Origin':'first', 'Dest':'last'}).sort_values(['ArrDelay','DepDelay','Origin','Dest'], ascending=False).reset_index(drop=True)
vuelos_delay
```

### 5.3 Busca las rutas más largas

Aquí también se usaron `.sort_values()` y `.drop_duplicates()`

```python
rutas_largas = df_flights_copia[['Origin', 'Dest', 'Distance']].drop_duplicates().sort_values(by='Distance', ascending=False).reset_index(drop=True)
rutas_largas
```

### 5.4 Y las que acumulen más retardos

```python
df_flights_copia.groupby(['Origin', 'Dest'])[['ArrDelay', 'DepDelay']].agg('sum').drop_duplicates().sort_values(by='ArrDelay', ascending=False)
```
